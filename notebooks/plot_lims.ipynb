## Plotting Grain Size Data with New Depth Limit for SR19-P4

Since XRF and micropaleontolgical data exist for SR19-P4 only to depth (in core) of 263 cm (from the top of the core down), there's no need to plot grain size data further down beyond that depth limit.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from grainsize import GrainSize
```

Import data for SR19-P4

```python
sr19_data = pd.read_csv(".../dataa-files/gs-files/sr19-stats-fines-corrected.csv", index_col=0)
```

```python
# sr19_data
```

Now slice `sr19` df so that it will only contain values down to depth of 263 cm and coerce the result to a `GrainSize` object:

```python
sr19_data = sr19_data[sr19_data.index <= 263]
sr19 = GrainSize(dataframe=sr19_data)
```

```python
# sr19
```

Now plot the grain size statistics

```python
plot_path = "../data-files/gs-files/sr19-stats-261.png"
sr19_plot = sr19.plot_stats_fines(fine_col="total", core_name="SR19-P4", marker=".", figsize=(10, 8), save_fig=False, fpath=plot_path)
```

In order to ensure that the plot for SR21-P7 is similar, we'll plot it here again.

```python
sr21_data = pd.read_csv("../data-files/gs-files/sr21-stats-fines-corrected.csv", index_col=0)
sr21 = GrainSize(dataframe=sr21_data)
sr21_plot_path = "../data-files/gs-files/sr21-stats-234.png"
```

```python
sr21_plot = sr21.plot_stats_fines(fine_col="total", core_name="SR21-P7", marker=".", figsize=(10, 8), save_fig=False, fpath=sr21_plot_path)
```

### A Comparison of SR19-P4 and SR21-P7 - Combined Grain Size Plot

```python
gs_comparison_plot = GrainSize.compare_gs(
    [sr19, sr21],
    core_names=["SR19-P4", "SR21-P7"],
    save_fig=False,
    fpath="../data-files/gs-files/gs-comparison.png",
)
```
