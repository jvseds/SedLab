### Combined Plots of Grain Size and XRF

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from grainsize import GrainSize, XRF, CoreAnalysis
```

#### SR19-P4

Import the grain size and XRF data

```python
sr19_gs = pd.read_csv("./data-files/gs-files/sr19-stats-fines-corrected.csv", index_col=0)
sr19_gs = GrainSize(dataframe=sr19_gs[sr19_gs.index <= 263])
```

```python
# define division factor
factor = 10_000
```

```python
sr19_xrf = pd.read_csv("./data-files/xrf-files/sr19_ppm.csv", index_col=0)
sr19_xrf = XRF(dataframe=(sr19_xrf.loc[:, :] / factor))
```

We'll plot the elements `Mg` and `Fe` and the statistics `median`, `mode` and `mean`.

```python
selected_els = ["Mg", "Fe"]
selected_stats = ["median", "mode", "mean"]
```

Plot for SR19

```python
sr19_combined = CoreAnalysis.plot_combined_stats(sr19_gs, sr19_xrf, core_name="SR19P4", xrf_els=selected_els)
```
